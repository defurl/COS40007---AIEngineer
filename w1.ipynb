#### Activity 2: Data exploration

```python
# Library Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# --- 1. Understanding the Data ---

try:
    data = pd.read_csv('concrete.csv')
    print("Data loaded successfully.")
except FileNotFoundError:
    print("Error: 'concrete.csv' file not found.")

print("First 5 rows of the dataset:")
print(data.head(5))

print("\nDataset Info:")
data.info()

print(f"\nInitial dataset shape: {data.shape[0]} rows, {data.shape[1]} columns")
```

```python
# --- 2. Data Cleaning ---

data.drop_duplicates(inplace=True)
print(f"Dataset shape after removing duplicates: {data.shape[0]} rows, {data.shape[1]} columns")
```

```python
print("\nGenerating boxplots to visualize outliers before removal...")
plt.figure(figsize=(15, 10))
col_num = data.shape[1]
for i in range(col_num):
    plt.subplot(3, 3, i + 1)
    plt.boxplot(data.iloc[:, i])
    plt.title(data.columns[i])
    plt.ylabel("Value")
    plt.xticks(rotation=45)
    plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print("\nRemoving outliers based on the IQR method...")
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

# Define the outlier boundary
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the dataframe to remove outliers
data_cleaned = data[~((data < lower_bound) | (data > upper_bound)).any(axis=1)]

print(f"Dataset shape after removing outliers: {data_cleaned.shape[0]} rows, {data_cleaned.shape[1]} columns")
print(f"Number of rows removed as outliers: {data.shape[0] - data_cleaned.shape[0]}")

plt.figure(figsize=(15, 10))
col_num = data.shape[1]
for i in range(col_num):
    plt.subplot(3, 3, i + 1)
    plt.boxplot(data.iloc[:, i])
    plt.title(data.columns[i])
    plt.ylabel("Value")
    plt.xticks(rotation=45)
    plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print("Missing values in the cleaned data:")
print(data_cleaned.isnull().sum())
```

#### Activity 3: EDA

```python
print("--- Starting Exploratory Data Analysis ---")

# Identify Target and Predictors
target_variable = 'strength'
predictors = data_cleaned.columns.drop(target_variable)

print(f"\nTarget Variable: {target_variable}")
print(f"Predictor Variables: {list(predictors)}")
```

```python
# Univariate Analysis: Generate histograms for each variable
print("\nGenerating histograms for Univariate Analysis...")
data_cleaned.hist(figsize=(20, 15), bins=30, edgecolor='black')
plt.suptitle('Distribution of Each Variable in the Concrete Dataset', size=20)
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```

```python
# Summary Statistics
print("\n--- Summary Statistics of the Cleaned Dataset ---")
summary_stats = data_cleaned.describe()
print(summary_stats)
```

```python
# Multivariate Analysis: Correlation Matrix and Heatmap
print("\n--- Performing Multivariate Analysis ---")
correlation_matrix = data_cleaned.corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(12, 10))
cax = plt.matshow(correlation_matrix, fignum=plt.gcf().number, cmap='viridis')
plt.colorbar(cax)

plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)

# Add correlation values to the heatmap
for i, col_i in enumerate(correlation_matrix.columns):
    for j, col_j in enumerate(correlation_matrix.columns):
        value = correlation_matrix.loc[col_i, col_j]
        color = 'white' if abs(value) > 0.5 else 'black'
        plt.text(j, i, f'{value:.2f}', ha='center', va='center', color=color)

plt.title('Correlation Matrix of Concrete Components', pad=80)
plt.show()
print("\n--- Exploratory Data Analysis Complete ---")
```
