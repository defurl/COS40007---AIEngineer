```python
import os
import glob
os.chdir('/content/drive/My Drive/Colab Notebooks/ampc')

csv_files = glob.glob('*.{}'.format('csv'))
print(csv_files)
```

```python
import pandas as pd
df_csv_concat = pd.concat([pd.read_csv(file) for file in csv_files], ignore_index=True)
df_csv_concat.head()
```

```python
df_csv_concat.to_csv('combined_data.csv', index=False)
shuffled_df = df_csv_concat.sample(frac=1).reset_index(drop=True)
shuffled_df.head()
```

```python
shuffled_df.to_csv('all_data.csv', index=False)
```

```python
all_df = pd.read_csv('all_data.csv')
all_df.head()
```

```python
# split target variabe and features
X = all_df.drop('class', axis=1)
y = all_df['class']

print("Shape of X: ", X.shape)
print("Shape of y: ", y.shape)
```

```python
# train svm
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
clf_73 = svm.SVC()
clf_73.fit(X_train, y_train)
y_pred = clf_73.predict(X_test)
print("Accuracy: ", accuracy_score(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
clf_cross = svm.SVC()
scores = cross_val_score(clf_cross, X, y, cv=5)
print("Cross Validation Scores: ", scores)
```

```python
# activity 3

from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.1, 1, 10],
              'gamma': [1, 0.1, 00.1],
              'kernel': ['rbf']}
grid_73 = GridSearchCV(clf_73, param_grid, refit=True, verbose=3)
grid_73.fit(X, y)
print("Best Parameters: ", grid_73.best_params_)
print("Best Score: ", grid_73.best_score_)
```

```python
#actitiyv 4 feature selection

from sklearn.feature_selection import SelectKBest, f_classif
selector = SelectKBest(f_classif, k=100)
X_new = selector.fit_transform(X, y)
print("Shape of X_new: ", X_new.shape)
```

```python
X_new_train, X_new_test, y_train, y_test = train_test_split(X_new, y, test_size=0.3, random_state=42)
grid_73.fit(X_new_train, y_train)
y_pred = grid_73.predict(X_new_test)
print("Accuracy: ", accuracy_score(y_test, y_pred))
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
X_pca = pca.fit_transform(X)
print("Shape of X_pca: ", X_pca.shape)
```

```python
X_pca_train, X_pca_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.4, random_state=42)
grid_73.fit(X_pca_train, y_train)
y_pred = grid_73.predict(X_pca_test)
print("Accuracy: ", accuracy_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
clf_rf = RandomForestClassifier(n_estimators=100, random_state=42)
clf_rf.fit(X_train, y_train)
y_pred_rf = clf_rf.predict(X_test)
print("RandomForest Accuracy: ", accuracy_score(y_test, y_pred_rf))
```

```python
# stochastic gradient descent
from sklearn.linear_model import SGDClassifier
clf_sgd = SGDClassifier(max_iter=1000, tol=1e-3, random_state=42)
clf_sgd.fit(X_train, y_train)
y_pred_sgd = clf_sgd.predict(X_test)
print("SGD Accuracy: ", accuracy_score(y_test, y_pred_sgd))
```

```python
# nn
from sklearn.neural_network import MLPClassifier
clf_nn = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)
clf_nn.fit(X_train, y_train)
y_pred_nn = clf_nn.predict(X_test)
print("NN Accuracy: ", accuracy_score(y_test, y_pred_nn))
```
